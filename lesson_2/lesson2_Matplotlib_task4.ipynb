```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
DATASET_PATH = '/home/valentine/Documents/GU_AI_396/03_ Data Science Python libs: Numpy, Matplotlib, Scikit-learn/homeworks/data/creditcard.csv'
```

```python
df = pd.read_csv(DATASET_PATH)
df.info()
```

```python
df.describe()
```

```python
df['Class'].value_counts()
```

```python
df['Class'].value_counts().plot(kind='bar')
```

```python
df['Class'].value_counts().plot(kind='bar', logy=True)
```

```python
df_v1 = df[['V1', 'Class']]
df_v1
```

```python
plt.hist(df_v1[df['Class'] == 1]['V1'], color='red', bins=20, alpha=0.5, density=True, label='Class 1')
plt.hist(df_v1[df['Class'] == 0]['V1'], color='gray', bins=20, alpha=0.5, density=True, label='Class 0')

plt.axis([-34, 4, 0, 0.24])
plt.xlabel('Class')
plt.legend()

plt.show()
```
